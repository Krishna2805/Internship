```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from scipy.signal import convolve2d
```

```python
(x_train,y_train), _ = mnist.load_data()
```

```python
image = x_train[10]
label = y_train[10]

plt.imshow(image,cmap='gray')
plt.title(f'Lbael: {label}')
plt.show()
```

```python
vertical_kernel = np.array([[-1,1,-1],
                            [-1,1,-1],
                            [-1,1,-1]])

horizontal_kernel = np.array([[-1,-1,-1],
                            [1,1,1],
                            [-1,-1,-1]])

corner_kernel = np.array([[-1,-1,-1],
                            [-1,1,1],
                            [-1,1,1]])
```

```python
vertical_output = convolve2d(image,vertical_kernel, mode='same', boundary='fill', fillvalue=0)
horizontal_output = convolve2d(image,horizontal_kernel, mode='same', boundary='fill', fillvalue=0)
corner_output = convolve2d(image,corner_kernel, mode='same', boundary='fill', fillvalue=0)

plt.figure(figsize=(10,5))
titles = ["Original Image", "Vertical Line Detector", "Horizontal Line Detector", "Corner Detector"]
images = [image, vertical_output, horizontal_output, corner_output]

for i in range(4):
    plt.subplot(1,4,i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
