```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
```

```python
model_name = "facebook/nllb-200-distilled-600M"
tokenizer = model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
input_text = "Hello, I’m Krishna, I am 20 years old and I am studying Computer Science."
```

```python
def translate(text, source_lang="eng_Latn", target_lang="kan_Knda"):
    translator = pipeline(
        "translation",
        model=model,
        tokenizer=tokenizer,
        src_lang=source_lang,
        tgt_lang=target_lang,
        max_length=400
    )
    return translator(text)[0]['translation_text']
```

```python
kannada_translation  = translate(input_text, target_lang="kan_Knda")
tamil_translation    = translate(input_text, target_lang="tam_Taml")
japanese_translation = translate(input_text, target_lang="jpn_Jpan")
french_translation   = translate(input_text, target_lang="fra_Latn")
```

```python
print(f"Kannada  : {kannada_translation}")
print(f"Tamil    : {tamil_translation}")
print(f"Japanese : {japanese_translation}")
print(f"French   : {french_translation}")
```
